# 1. Read the dataset to the python environment

```python
import pandas as pd
data = pd.read_csv('cars_data.csv')
```

```python
data
```

# 2. Check for the null values present in the dataset

```python
data.isnull().sum()
```

Hence null values present in the dataset is zero.

# 3. Bar graph of male vs female buyers

```python
df = pd.DataFrame(data)
```

```python
df["Buyer Gender"].value_counts()
```

```python
df["Buyer Gender"].value_counts().plot(kind='bar')
import matplotlib.pyplot as plt
plt.title("Bar graph of male buyers vs female buyers",fontsize=15)
plt.xlabel("Buyer Gender")
plt.ylabel("Number of Buyers")
plt.show()
```

Hence the graph is displayed with male buyers 4948 and female buyers 5052.

# 4. Top 5 cars based on their sales price

```python
top_5 = df.sort_values("Sale Price",ascending=False)
top_5.head(5)
```

Hence top 5 cars based on their sales price are Volkswagen, Audi, Lexus LX, Mazda and Lexus ES.

# 5. Least 5 cars based on their Resell price

```python
bott_5 = df.sort_values("Resell Price",ascending=True)
bott_5.head(5)
```

Hence least 5 cars based on their resell price are Chevrolet, GMC, Jaguar, Mitsubishi and Infiniti.
